```python
import mne
import numpy
import matplotlib.pyplot as plt
from igraph import *
def load_data():

    base_path = 'D:/Physionet_preprocessing/S%03dR%02d.set'
    subjects = 104
    train_data = numpy.zeros((104 * 42, 64, 657))
    train_label = numpy.zeros((104 * 42))

    MI_tasks_4_class = [4,6,8,10,12,14]
    MI_tasks_2_class = [4,8,12]

    total_id = 0

    for subject_id in range(1, subjects + 6):
        for MI_task in MI_tasks_2_class:
            if subject_id not in [88, 89, 92, 100,104]:
                path = base_path % (subject_id, MI_task)
                print(path)

                raw = mne.io.read_raw_eeglab(path,preload = True)
                events,event_dict= mne.events_from_annotations(raw)
                picks = mne.pick_types(raw.info,meg=False,eeg=True,eog=False,stim=False)
                
                # Extracts epochs of 4.1s time period from the datset
                tmin,tmax = 0.,4.1
                event_id = dict({'T2':2,'T3':3})
                epochs = mne.Epochs(raw,events,event_id,tmin,tmax,proj=True,picks=picks,baseline=None,preload=True)
                
                data = epochs.get_data()[0:14] * 1e06

                if MI_task == 4 or MI_task == 8 or MI_task == 12:
                    labels = epochs.events[:,-1] -2
                    labels = labels[0:14]
                else:
                    labels = epochs.events[:,-1]
                    labels = labels[0:14]
                
                train_data[total_id:total_id + 14] = data.reshape((14, 64,657))
                train_label[total_id:total_id + 14] = labels
                total_id = total_id + 14

    print(train_data)
    print(train_label)
    return train_data,train_label


train_data,train_label = load_data()
# 计算皮尔逊矩阵
coff = numpy.corrcoef(train_data[0]) - numpy.identity(64)
for i in range(1,len(train_data)):
    coff += (numpy.corrcoef(train_data[i]) - numpy.identity(64))
coff = coff / len(train_data)
```

```python
vertexs = ['Fc5', 'Fc3', 'Fc1', 'Fcz', 'Fc2', 'Fc4', 'Fc6', 'C5', 'C3', 'C1', 'Cz', 'C2', 'C4',
                'C6', 'Cp5', 'Cp3', 'Cp1', 'Cpz', 'Cp2', 'Cp4', 'Cp6', 'Fp1', 'Fpz', 'Fp2', 'Af7', 'Af3',
                'Afz', 'Af4', 'Af8', 'F7', 'F5', 'F3', 'F1', 'Fz', 'F2', 'F4', 'F6', 'F8', 'Ft7', 
                'Ft8', 'T7', 'T8', 'T9', 'T10', 'Tp7', 'Tp8', 'P7', 'P5', 'P3', 'P1', 'Pz', 'P2', 
                'P4', 'P6', 'P8', 'Po7', 'Po3', 'Poz', 'Po4', 'Po8', 'O1', 'Oz', 'O2', 'Iz']


# 计算全连接的最小生成树
span_g = Graph()
span_g.add_vertices(64)
span_g.vs["name"] = vertexs

m=0
n=1
i=1
span_edge = []
span_coff_weight = []
for m in range(0,63):
    for i in range(n,64):
            span_edge.append((m,i))
            span_coff_weight.append(coff[m][i])   #加入边
    n = n + 1
    
span_g.add_edges(span_edge)
span_g.es["weight"] = span_coff_weight
# 生成最小生成树
span_tree = span_g.spanning_tree(weights=span_g.es["weight"])
```

```python
best_q = 0.0   #最好的模块度值
best_membership = []  #最好模块度的区域分布
best_coff = 0.0  # 最好模块的皮尔逊值
best_com = None  

# 记录每一个不一样的模块度的相关数据
threshold = []
q_score = []
memberships = []

# plot  X轴，Y轴
x_coffs = numpy.array([])
y_modularization_cof = numpy.array([])

# 以0.001为步长，获取皮尔逊连接
for mk in numpy.arange(0.0,0.9,0.001):
    # 构建图
    g = Graph()
    g.add_vertices(64)
    g.vs["name"] = vertexs
    
    m=0
    n=1
    i=1
    edge = []
    coff_weight = []
    for m in range(0,64):
        for i in range(n,64):
            if coff[m][n] >= mk:
                edge.append((m,i))
                coff_weight.append(coff[m][i])
            else:
                continue
        n = n + 1
    g.add_edges(edge)
    g.es["weight"] = coff_weight
    
    xx = g.es["weight"] # 将边的权重值复制
    
    for i in range(0,64):
        span_weight = 0
        best_weight_k = -1
        if g.neighbors(i) == []:
            span_neighbors = span_tree.neighbors(i)
            if(len(span_neighbors) == 1):
                span_weight = span_tree[(i,span_neighbors[0])]
                best_weight_k = span_neighbors[0]
            else:
                span_weight = span_tree[(i,span_neighbors[0])]
                best_weight_k = span_neighbors[0]
                for k in range(1,len(span_neighbors)):
                    if span_weight > span_tree[(i,span_neighbors[k])]:
                        span_weight = span_tree[(i,span_neighbors[k])]
                        best_weight_k = span_neighbors[k]
                    else:
                        continue
            
            g.add_edge(i,best_weight_k)
            g.es["weight"].append(coff[i][best_weight_k])
            xx.append(coff[i][best_weight_k])
        else:
            continue

    g.es["weight"] = xx
    gg = g.spanning_tree(weights=g.es["weight"])
    com = gg.community_infomap()
    
    if com.q not in q_score:
        threshold.append(mk)
        q_score.append(com.q)
        memberships.append(com.membership)
    
    if com.q>best_q:
        best_q = com.q
        best_membership = com.membership
        best_coff = mk
        best_com = com
    x_coffs = numpy.append(x_coffs,[mk])
    y_modularization_cof = numpy.append(y_modularization_cof,[com.q])

for i in range(0,len(threshold)):
    print("********",i)
    print("        ",threshold[i])
    print("        ",q_score[i])
    print("        ",memberships[i])
plt.scatter(x_coffs,y_modularization_cof,s=5)
plt.xlabel("Pearson coefficient")
plt.ylabel("Modularity(Q)")
plt.savefig(r"C:\Users\pc\Desktop\SCI_images\Physionet.png",bbox_inches='tight')
plt.show()
    
print("best_membership:")
print(best_membership)
print("best_q:")
print(best_q)
print("best_coff:")
print(best_coff)

plot(best_com)
```
